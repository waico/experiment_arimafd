```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['figure.figsize'] = 8,4
```

```python
def sigm_scale(x,A_tp,A_fp,window=1):
    y = (A_tp-A_fp)*(1/(1+np.exp(5*x/window))) + A_fp
    y[:len(x[x<-3])] = A_fp
    return y 
```

```python
x = np.linspace(-4,3,400)
x1 = np.linspace(-1,6,400)
```

```python
plt.axvline(4,c='b',ls='--')
plt.axvline(-0,c='b',ls='--')
plt.axhline(1,c='b',ls='--')
plt.axhline(-1,c='b',ls='--')
plt.plot(x1,sigm_scale(x,1,-1),linewidth=4.1,c='black')
plt.grid()
plt.ylim(-1.4,1.4)
plt.yticks([1,0.0,-1],labels=['$A_{tp}$',0.0,'$A_{fp}$'],size=12)
plt.xticks(size=12)
plt.xlabel('Relative position',size=12)
plt.ylabel('Value of sigmoid \n scoring function',size=12)
plt.savefig(fname='sigm_nab', fmt='png', dpi=150)
```

```python
pd.DataFrame(sigm_scale(x,1,-1), index=x,columns=['Sigmoid scoring function']).to_csv('sigm_nab.csv')
```
