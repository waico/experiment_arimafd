```python
x=list(np.arange(-2,2,0.01))
```

```python
def diff_huber(x,delta):
    answer = np.array([])
    for i in x:
        if np.abs(i) < delta:
            answer = np.append(answer,i)
        else:
            answer = np.append(answer,delta*np.sign(i) - 1/2*delta**2)
    return answer
        
```

```python
nnn={1:'MSE',2:'MAE',3:'log-cosh',4:'Huber, $ \delta=1 $',5:'Huber, $ \delta=0.1 $',6:'Huber, $ \delta=10 $'}
```

```python
plt.figure(figsize=(9, 8))
# plt.subplots_adjust(wspace=0, hspace=0.4)
plt.subplot(211)

plt.plot(x,np.tanh(x),'o',ms=1,label=nnn[3])
plt.plot(x,np.sign(x),'o',ms=1,label=nnn[2])
plt.plot(x,x,'o',ms=1,label=nnn[1])
plt.legend()

plt.ylabel('Значение производной',fontsize=12)
plt.xlabel('Значение аргумента',fontsize=12)
plt.ylim(-2,2)

plt.grid()
# plt.arrow(-5,0,10,0)
# plt.arrow(0,-5,0,10)


ax = plt.subplot(212)

plt.plot(x,diff_huber(x,0.1),'o',ms=1,label=nnn[4])
plt.plot(x,diff_huber(x,1),'o',ms=1,label=nnn[5])
plt.plot(x,diff_huber(x,0.5),'o',ms=1,label=nnn[6])
plt.grid()
plt.ylim(-2,2)
plt.legend()

plt.ylabel('Значение производной',fontsize=12)
plt.xlabel('Значение аргумента',fontsize=12)
save('loss')
```
