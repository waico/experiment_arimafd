```python
def save(name='', fmt='png'):
    pwd = os.getcwd()
    iPath = pwd + '\\pictures'
    if not os.path.exists(iPath):
        os.mkdir(iPath)
    os.chdir(iPath)
    plt.savefig('{}.{}'.format(name, fmt), fmt='png', dpi=150, bbox_inches='tight')
    os.chdir(pwd)
```

```python
path='C:/data/'
os.getcwd()
data=pd.read_csv(path+'case2.csv',index_col='Unnamed: 0',parse_dates=True,encoding='cp1251')
data1=pd.DataFrame()
for i in data:
    if ('вибрац' in i.lower()) and ('11 ' in i.lower()):
        data1[i]=data[i]
print(data1.columns)
sig1=data1.iloc[:,0].dropna().resample('T').mean().interpolate().fillna(method='bfill').fillna(method='ffill').dropna()
sig2=data1.iloc[:,1].dropna().resample('T').mean().interpolate().fillna(method='bfill').fillna(method='ffill').dropna()
sig3=data1.iloc[:,2].dropna().resample('T').mean().interpolate().fillna(method='bfill').fillna(method='ffill').dropna()
signal=np.sqrt(sig1**2+sig2**2+sig3**2)
# train=signal['2017-05-01':'2017-05-10']
# test=signal['2017-05-11':]
# accident='2017-05-17 '
```

### MAPE

```python
def mape(df_true,df_fore):
    df = (df_true - df_fore)/df_true
    return df.abs().mean()*100
tt = []
for i in res:
    mape(test1,dat[i]).plot(label=name[i])
    tt.append(mape(test1,dat[i]))
    plt.legend()
    plt.grid()
plt.xlabel('Forecasting point from the end of the dataset',fontsize=12)
plt.ylabel('Mean absolute percentage error,%',fontsize=12)
save('errors')
plt.show()
```

```python
df = pd.concat(tt,axis=1)
df.columns = [i for i in name.values()]
df.index = df.index + 1
df.to_csv('errors.csv',sep=';')
```
